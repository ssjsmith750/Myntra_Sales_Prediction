```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys


```

```python

driver = webdriver.Chrome()

driver.get('https://www.myntra.com/men-jackets')

ElementBoxes = driver.find_elements(By.XPATH,"//li[@class='product-base']")
Brand = driver.find_elements(By.XPATH,"//h3[@class='product-brand']")
Discounted_Price = driver.find_elements(By.XPATH,"//span[@class='product-discountedPrice']")
Actual_price = driver.find_elements(By.XPATH,"//span[@class='product-strike']")
Discount_percentage = driver.find_elements(By.XPATH,"//span[@class='product-discountPercentage']")
Image_Element = driver.find_element(By.XPATH,"//div[@class='product-imageSliderContainer']")
Links = driver.find_elements(By.XPATH,"//a[@class='product-base']")

time.sleep(3)
```

```python
Details = {'Brand':[],'Description':[],'Discounted_Price':[],'Actual_price':[],'Discount_percentage':[],'Links':[]}
```

```python
driver = webdriver.Chrome()

#driver.get('https://www.myntra.com/men-jackets')
driver.get('https://www.myntra.com/footwear?src=bc')

driver.maximize_window()

time.sleep(3)

for i in range(300):
    ElementBoxes = driver.find_elements(By.XPATH,"//li[@class='product-base']")
    #ElementBoxes = driver.find_elements(By.XPATH,"//li[@class='product-base']")

    for i in ElementBoxes:

        try:brand = i.find_element(By.XPATH,"//h3[@class='product-brand']").text
        except:brand = None

        try:discounted_Price = i.find_element(By.XPATH,"//span[@class='product-discountedPrice']").text
        except:discounted_Price = None

        try:actual_price = i.find_element(By.XPATH,"//span[@class='product-strike']").text
        except:actual_price = None

        try:discount_percentage = i.find_element(By.XPATH,"//span[@class='product-discountPercentage']").text
        except:discount_percentage = None

        #try:image_Element = i.find_element(By.CSS_SELECTOR,"img.img-responsive").get_attribute('src')
        #except:image_Element = None

        try:links = i.find_element(By.CSS_SELECTOR,"a[data-refreshpage='true'").get_attribute('href')
        except:links = None

        try:description = i.find_element(By.XPATH,"//h4[@class='product-product']").text
        except:description = None

        Details['Brand'].append(brand)
        Details['Discounted_Price'].append(discounted_Price)
        Details['Actual_price'].append(actual_price)
        Details['Discount_percentage'].append(discount_percentage)
        #Details['Image_Element'].append(image_Element)
        Details['Links'].append(links)
        Details['Description'].append(description)

    driver.find_element(By.XPATH,"//span[@class='pagination-arrowRight']").click()
    time.sleep(3)
    
```

```python
for i in Details.keys():
    print(i,len(Details[i]))
```
